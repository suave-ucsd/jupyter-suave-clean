# Adding concepts based on content of images. They are added as tags#multi to the dataset

```python
import os
from clarifai.rest import ClarifaiApp

import pandas as pd
import sys
# these are only for Python 2
# reload(sys)
# sys.setdefaultencoding('utf-8')
import glob, os
```

```python
api_key = ""

if api_key == '':
    print('Cannot proceed without an API Key')
else:
    os.environ["CLARIFAI_API_KEY"] = api_key

    print("Your CLARIFAI App Key is:", os.environ["CLARIFAI_API_KEY"])
    app = ClarifaiApp()
    model = app.public_models.general_model
```

```python
# Input files: 
# either a folder with images or a file list or a url list

list_in_file = False
URL_or_local = 'local'

if list_in_file:
    inputfile = "C:\\Users\\izaslavsky\\Downloads\\surveys\\MOPA\\all_366_images_small\\allfiles_list.csv"

    set_of_files = []
    with open(inputfile) as f:
        set_of_files = f.read().splitlines() 
else:
    os.chdir("C:\\Users\\izaslavsky\\Downloads\\surveys\\MOPA\\MOPA_med_images_photoswipe")
    set_of_files = glob.glob("*.*")

print("Count of items to process: ", len(set_of_files))

all_data = []

outputfile = "C:\\Users\\izaslavsky\\Downloads\\surveys\\allconcepts.csv"
```

```python
# Run the model
for image in set_of_files:
    if URL_or_local == 'url':
        response = model.predict_by_url(url=image)
    else:
        response = model.predict_by_filename(filename=image)        
    file_data = {}
    file_data['image_name'] = image
    print('Current image: ',image)
    concepts = response['outputs'][0]['data']['concepts']
    for i in range(len(concepts)):  
        print(concepts[i]['name'], concepts[i]['value'])
        file_data['concept_'+str(i+1)] = concepts[i]['name']
        file_data['value_'+str(i+1)] = concepts[i]['value']
    all_data.append(file_data)
```

```python
df = pd.DataFrame(all_data).fillna('')
df = df[['image_name', 
         'concept_1', 'value_1',
         'concept_2', 'value_2',
         'concept_3', 'value_3',
         'concept_4', 'value_4',
         'concept_5', 'value_5',
         'concept_6', 'value_6',
         'concept_7', 'value_7',
         'concept_8', 'value_8',
         'concept_9', 'value_9',
         'concept_10', 'value_10',
         'concept_11', 'value_11',
         'concept_12', 'value_12',
         'concept_13', 'value_13',
         'concept_14', 'value_14',
         'concept_15', 'value_15',
         'concept_16', 'value_16',
         'concept_17', 'value_17',
         'concept_18', 'value_18',
         'concept_19', 'value_19',
         'concept_20', 'value_20']]
multi =''
# multi = df['concept_'+str(1)]
for i in range(20):
    if i == 19:
        multi += df['concept_'+str(i+1)]
    else:
        multi += df['concept_'+str(i+1)] +'|'

df['tags#multi'] = multi
df.head()['tags#multi']
```

```python
# new filename
df.to_csv(outputfile, index=None)

# display a URL for the new survey
```
